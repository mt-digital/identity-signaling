```python
disliking_df = pd.read_csv('../data/invasion_beta1e6/disliking/full.csv')
recept_df = pd.read_csv('../data/invasion_beta1e6/receptivity/full.csv')
```

```python
figdir = '../../papers/idsig/Figures/'
```

```python
# Double checking number of trials in each.
df = disliking_df
df = df[(df.disliking == 0.1) & (df.homophily == 0.1) 
        & (df.initial_prop_churlish == 0.05) & (df.initial_prop_covert == 0.05)]
print(len(df.trial_idx.unique()))

df = recept_df
df = df[(df.receptivity == 0.1) & (df.homophily == 0.1) 
        & (df.initial_prop_churlish == 0.05) & (df.initial_prop_covert == 0.05)]

print(len(df.trial_idx.unique()))
```
