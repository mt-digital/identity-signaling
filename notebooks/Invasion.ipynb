```python
import pandas as pd
import warnings; warnings.simplefilter('ignore')
import sys
sys.path.append('..')
from id_signaling.figures import invasion_heatmaps
```

## $\beta = 10$

```python
disliking_df = pd.read_csv('../data/invasion/disliking/full.csv')
recept_df = pd.read_csv('../data/invasion/receptivity/full.csv')
```

```python
figdir = '../../papers/idsig/Figures/'
```

```python
invading_prev = 0.05
# Double checking number of trials in each.
df = disliking_df
df = df[(df.disliking == 0.2) & (df.homophily == 0.1) 
        & (df.initial_prop_churlish == invading_prev) & (df.initial_prop_covert == invading_prev)]
print(len(df.trial_idx.unique()))

df = recept_df
df = df[(df.receptivity == 0.2) & (df.homophily == 0.1) 
        & (df.initial_prop_churlish == invading_prev) & (df.initial_prop_covert == invading_prev)]

print(len(df.trial_idx.unique()))
```

```python
# print(disliking_df.disliking.unique())
disliking_df.head()
```

```python
invasion_heatmaps(disliking_df, recept_df, invading='covert', timesteps=100,
                  save_path=f'{figdir}/covert_invades.pdf', annot=True, 
                  figsize=(10, 7),
                  invading_prev=invading_prev, 
                  cbar_label_size=13);
```

```python
invasion_heatmaps(disliking_df, recept_df, invading='overt',
                  save_path=f'{figdir}/overt_invades.pdf', invading_prev=invading_prev,
                  timesteps=100, annot=True);
```

```python
invasion_heatmaps(disliking_df, recept_df, invading='churlish',
                  save_path=f'{figdir}/churlish_invades.pdf', invading_prev=invading_prev,
                  timesteps=100, annot=True, cbar_label_size=12.5);
```

```python
invasion_heatmaps(disliking_df, recept_df, invading='generous',
                  save_path=f'{figdir}/generous_invades.pdf', invading_prev=invading_prev,
                  timesteps=100, annot=True);
```
