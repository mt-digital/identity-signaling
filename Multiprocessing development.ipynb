```python
from id_signaling.experiment import (
    analyze_covert_receiving_prob, vary_covert_receiving_prob
)

res = vary_covert_receiving_prob(
    covert_rec_probs=[0.05, 0.25, 0.45], 
    n_trials=10,
    similarity_benefit=0.4
)
```
